```python
# Name: M. Saad Bin Shafiq
# Reg No: 200901079
# AI Assignment 2

import sys

def is_terminal(board):
    # Check for a winner
    for row in range(3):
        if board[row][0] == board[row][1] == board[row][2] and board[row][0] != " ":
            return True
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col] and board[0][col] != " ":
            return True
    if board[0][0] == board[1][1] == board[2][2] and board[0][0] != " ":
        return True
    if board[0][2] == board[1][1] == board[2][0] and board[0][2] != " ":
        return True

    # Check for a draw
    for row in board:
        for cell in row:
            if cell == " ":
                return False
    return True

def utility(board):
    for row in range(3):
        if board[row][0] == board[row][1] == board[row][2] and board[row][0] != " ":
            return 1 if board[row][0] == "X" else -1
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col] and board[0][col] != " ":
            return 1 if board[0][col] == "X" else -1
    if board[0][0] == board[1][1] == board[2][2] and board[0][0] != " ":
        return 1 if board[0][0] == "X" else -1
    if board[0][2] == board[1][1] == board[2][0] and board[0][2] != " ":
        return 1 if board[0][2] == "X" else -1
    return 0

def min_value(board):
    if is_terminal(board):
        return utility(board)
    v = sys.maxsize
    for row in range(3):
        for col in range(3):
            if board[row][col] == " ":
                board[row][col] = "O"
                v = min(v, max_value(board))
                board[row][col] = " "
    return v

def max_value(board):
    if is_terminal(board):
        return utility(board)
    v = -sys.maxsize
    for row in range(3):
        for col in range(3):
            if board[row][col] == " ":
                board[row][col] = "X"
                v = max(v, min_value(board))
                board[row][col] = " "
    return v

def minmax(board):
    best_move = (-1, -1)
    best_value = -sys.maxsize
    for row in range(3):
        for col in range(3):
            if board[row][col] == " ":
                board[row][col] = "X"
                move_value = min_value(board)
                board[row][col] = " "
                if move_value > best_value:
                    best_value = move_value
                    best_move = (row, col)
    return best_move

def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("---------")

def main():
    board = [[" " for _ in range(3)] for _ in range(3)]
    
    print("                       Tic Tac Toe Game                          ")
    print("*************   This Game consists of two Players   *************\n")
    Player1 = input("Enter the name of the First player: ")
    Player2 = input("Enter the name of the Second player: ")
    Players = {Player1: "X", Player2: "O"}
    current_player = Player1

    while not is_terminal(board):
        print_board(board)
        row = int(input(f"{current_player}, select the row from (0-2) in which you want to place your mark: "))
        col = int(input(f"{current_player}, select the column from (0-2) in which you want to place your mark: "))
        
        if board[row][col] == " ":
            board[row][col] = Players[current_player]
            current_player = Player2 if current_player == Player1 else Player1
        else:
            print("Invalid move. Please Try again.")
            continue

            
    print_board(board)

    result = utility(board)
    if result == 0:
        print("It's a draw!")
    elif result == 1:
        print(f"{Player1} wins!")
    else:
        print(f"{Player2} wins!")

if __name__ == "__main__":
    main()
```
